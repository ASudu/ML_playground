```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from id_6_K_means_cluster import k_means_clustering
import numpy as np
import random
```

```python
# Generate random sample blobs
k = 4
sample_points = make_blobs(n_samples=5000, centers=k, n_features=2)[0]

# Plot the sample points
x_coords, y_coords = zip(*sample_points)
plt.scatter(x_coords, y_coords)
plt.title("Sample Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
initial_centroids = random.sample(list(sample_points), k)
centroids, clus_map = k_means_clustering(sample_points, k, initial_centroids, max_iterations=5000)
```

```python
# Plot the clustered points with centroids
colors = ['r', 'g', 'b', 'c', 'm', 'y']

for i in range(k):
    clus_points = clus_map.get(i, [])
    if clus_points:
        x_coords, y_coords = zip(*clus_points)
        plt.scatter(x_coords, y_coords, color=colors[i % len(colors)], label=f'Cluster {i}')
        # color of centroid is "red"
        plt.scatter(centroids[i][0], centroids[i][1], color='k', marker='o', s=100)
```
